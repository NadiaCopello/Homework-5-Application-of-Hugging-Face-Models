```python
# Importar librerías
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
from datasets import load_dataset
import pandas as pd
```

```python
# Cargar modelo y tokenizer FinBERT
model_name = "ProsusAI/finbert"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
# Crear pipeline de análisis de sentimientos
finbert = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
#Cargar el conjunto de datos Financial Phrasebank desde Hugging Face
from datasets import load_dataset
dataset = load_dataset("financial_phrasebank", "sentences_allagree")
```

```python
# Convertir a DataFrame para fácil manipulación
df = pd.DataFrame(dataset['train'])
```

```python
# Mostrar algunas frases originales
print("Primeras frases del dataset:")
print(df.head())

# Paso 6: Aplicar FinBERT a una muestra de frases (puedes usar todas si deseas)
muestra = df['sentence'][:50]
```

```python
# Analizar sentimiento
resultados = finbert(muestra.tolist())

# Agregar resultados al DataFrame
df_resultados = pd.DataFrame(resultados)
df_muestra = pd.DataFrame(muestra)
df_muestra.reset_index(drop=True, inplace=True)
df_resultados = pd.concat([df_muestra, df_resultados], axis=1)
df_resultados.columns = ['Frase', 'Sentimiento', 'Score']

# Mostrar resultados
print("\nFrases con sentimiento estimado por FinBERT:")
print(df_resultados)
```

```python
# Subir el archivo PDF
from google.colab import files
uploaded = files.upload()
```

```python
#  Extraer texto del PDF usando PyMuPDF
import fitz  # pymupdf

filename = list(uploaded.keys())[0]
doc = fitz.open(filename)
texto = ""
for page in doc:
    texto += page.get_text()
```

```python
# Verificamos los primeros caracteres
print(texto[:1000])
```

```python
from transformers import pipeline
```

```python
# Crear pipeline de resumen
resumidor = pipeline("summarization", model="facebook/bart-large-cnn")

# Cortar el texto si es muy largo (modelo tiene límite de tokens)
# Ajusta el truncado si deseas más contexto
entrada = texto[:1024]

# Generar resumen
resumen = resumidor(entrada, max_length=500, min_length=100, do_sample=False)

# Mostrar resumen
print(" Resumen generado:\n")
print(resumen[0]['summary_text'])
```
